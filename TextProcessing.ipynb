```python
import pandas as pd
```

```python
training_data = pd.read_csv('/Users/vaishaliyadav/Practical_Mtech/TPDataSet.csv', sep=',', names=["sent", "class"])
training_data.head()
```

```python
prob_class=training_data.groupby('class').size().div(len(training_data))
print(prob_class)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

stmt_docs = [row['sent'] for index,row in training_data.iterrows() if row['class'] == 'statement']

vec_s = CountVectorizer()
X_s = vec_s.fit_transform(stmt_docs)
tdm_s = pd.DataFrame(X_s.toarray(), columns=vec_s.get_feature_names())

tdm_s.head()
```

```python
q_docs = [row['sent'] for index,row in training_data.iterrows() if row['class'] == 'question']

vec_q = CountVectorizer()
X_q = vec_q.fit_transform(q_docs)
tdm_q = pd.DataFrame(X_q.toarray(), columns=vec_q.get_feature_names())

tdm_q.head()
```

```python
c_docs = [row['sent'] for index,row in training_data.iterrows() if row['class'] == 'command']

vec_c = CountVectorizer()
X_c = vec_c.fit_transform(c_docs)
tdm_c = pd.DataFrame(X_c.toarray(), columns=vec_c.get_feature_names())

tdm_c.head()
```

```python
word_list_s = vec_s.get_feature_names();    
count_list_s = X_s.toarray().sum(axis=0) 
freq_s = dict(zip(word_list_s,count_list_s))
freq_s
```

```python
word_list_q = vec_q.get_feature_names();    
count_list_q = X_q.toarray().sum(axis=0) 
freq_q = dict(zip(word_list_q,count_list_q))
freq_q
```

```python
word_list_c = vec_c.get_feature_names();    
count_list_c = X_c.toarray().sum(axis=0) 
freq_c = dict(zip(word_list_c,count_list_c))
freq_c
```

```python
prob_s = []
for word, count in zip(word_list_s, count_list_s):
    prob_s.append(count/len(word_list_s))
dict(zip(word_list_s,prob_s))
```

```python
prob_q = []
for count in count_list_q:
    prob_q.append(count/len(word_list_q))
dict(zip(word_list_q,prob_q))
```

```python
prob_c = []
for count in count_list_c:
    prob_c.append(count/len(word_list_c))
dict(zip(word_list_c,prob_c))
```

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = [row['sent'] for index,row in training_data.iterrows()]

vec = CountVectorizer()
X = vec.fit_transform(docs)

total_features = len(vec.get_feature_names())
total_features
```

```python
total_cnts_features_s = count_list_s.sum(axis=0)
total_cnts_features_q = count_list_q.sum(axis=0)
total_cnts_features_c = count_list_c.sum(axis=0)

print(total_cnts_features_s)
print(total_cnts_features_q)
print(total_cnts_features_c)
```

```python
from nltk.tokenize import word_tokenize
new_sentence = 'When did the Dreyfus affair occur?'
new_word_list = word_tokenize(new_sentence)
```

```python
prob_s_with_ls = []
for word in new_word_list:
    if word in freq_s.keys():
        count = freq_s[word]
    else:
        count = 0
    prob_s_with_ls.append((count + 1)/(total_cnts_features_s + total_features))
dict(zip(new_word_list,prob_s_with_ls))
```

```python
prob_q_with_ls = []
for word in new_word_list:
    if word in freq_q.keys():
        count = freq_q[word]
    else:
        count = 0
    prob_q_with_ls.append((count + 1)/(total_cnts_features_q + total_features))
dict(zip(new_word_list,prob_q_with_ls))
```

```python
prob_c_with_ls = []
for word in new_word_list:
    if word in freq_c.keys():
        count = freq_c[word]
    else:
        count = 0
    prob_c_with_ls.append((count + 1)/(total_cnts_features_c + total_features))
dict(zip(new_word_list,prob_c_with_ls))
```

```python
import nltk
q=1
s=1
c=1
for i in range(len(new_word_list)):
    q=q*prob_q_with_ls[i]
    s=s*prob_s_with_ls[i]
    c=c*prob_c_with_ls[i]


if (((prob_class.iloc[1]*q) > (prob_class.iloc[2]*s)) and ((prob_class.iloc[1]*q) > (prob_class.iloc[0]*c))):
    print("Sentence is a Question")
elif (((prob_class.iloc[2]*s) > (prob_class.iloc[1]))*q and ((prob_class.iloc[2]*s) > (prob_class.iloc[0]*c))):
    print("Sentence is a Statement")
elif (((prob_class.iloc[0]*c) > (prob_class.iloc[1]*q)) and ((prob_class.iloc[0]*c) > (prob_class.iloc[2]*s))):
    print("Sentence is a Command")
```
